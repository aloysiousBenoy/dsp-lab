```python
x=np.arange(-3,5)
n=np.zeros_like(x)
def unitpulse(n,x,a):
    n[x>a]=1
    return n
plt.stem(x,unitpulse(n,x,-2))
plt.show()
```

```python
x=np.arange(-3,9)
n=np.zeros_like(x)
def ramp(n,x):
    n[(x>0)&(x<5)]=x[(x>0)&(x<5)]
    n[(x>4)&(x<9)]=8-x[(x>4)&(x<9)]
    return n
plt.stem(x,ramp(n,x))
plt.show()
```

```python
x=np.arange(-3,9)
n=np.zeros_like(x)
def bipolar(x,n):
    n[(x>=0)&(x<=2)]=1
    n[(x>=3)&(x<=5)]=-1
    return n
plt.stem(x,bipolar(x,n))
plt.xticks(x)
plt.show()
```

```python
#dft

def dft(x):
    N=4
    a=np.arange(N)
    w=np.exp(-1j*(2*np.pi)/N)
    d=np.empty((N,N),dtype=np.cdouble)
    for k in a:
        for n in a:
            d[k][n]=w**(n*k)        
    X=np.round(d@(x.T))
    return X
    
#print(d)

print(dft(np.array([1,2,3,4])))
```

```python
#idft

def idft(x):
    N=len(x)
    a=np.arange(N)
    w=np.exp(1j*(2*np.pi)/N)
    d=np.empty((N,N),dtype=np.cdouble)
    for k in a:
        for n in a:
            d[k][n]=w**(n*k)        
    X=np.round(d@(x.T))
    return X/N
    
#print(d)

print(idft(dft(np.array([1,2,3,4]))))
```

```python
def fft(u):
    u = np.array(u,dtype=complex)
    y=np.zeros_like(u)
    for i in np.arange(len(u)):
        y[i]=fft_k(u,i)
    return np.round(y)    
        
        
def fft_k(x,k):
    w=np.exp(-1j*2*np.pi*k/len(x))
    if(len(x)==1):
        return x[0]
    else:
        x_even=x[::2]
        x_odd=x[1::2]
        a=fft_k(x_even,k)
        b=fft_k(x_odd,k)
        return a+(w*b)
print(fft(np.array([1,2,3,4],dtype=complex)))    
        
        
    
```

```python
def ifft(u):
    y=np.zeros_like(u)
    for i in np.arange(len(u)):
        y[i]=ifft_k(u,i)
    return np.round(y)/len(u)    
        
        
def ifft_k(x,k):
    w=np.exp(1j*2*np.pi*k/len(x))
    if(len(x)==1):
        return x[0]
    else:
        x_even=x[::2]
        x_odd=x[1::2]
        a=ifft_k(x_even,k)
        b=ifft_k(x_odd,k)
        return a+(w*b)
print(ifft(np.array([10,-2+2j,-2,-2-2j],dtype=complex)))   
```
