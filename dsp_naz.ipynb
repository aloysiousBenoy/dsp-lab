```python

#BIPOLAR
 
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-3,10)
x=np.zeros_like(n)
 
x[(n>=0)&(n<=2)]=1
x[(n>=3)&(n<=5)]=-1
 
plt.stem(n,x,use_line_collection=True)
plt.show()
 
#PULSE
 
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-3,8)
x=np.zeros_like(n)
 
x[(n>=0)&(n<=4)]=1
 
plt.stem(n,x,use_line_collection=True)
plt.show()
 
 
 
 
#RAMP
 
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-3,5)
x=np.zeros_like(n)
 
x[(n>=0)&(n<=4)]=n[(n>=0)&(n<=4)]
 
 
plt.stem(n,x,use_line_collection=True)
plt.xticks(n) #to display each value in x axis
plt.show()
 
#SHIFT
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-2,9)
x=np.zeros_like(n)
 
x[(n>=3)&(n<=7)]=1
 
plt.stem(n,x)
plt.show()


#TRIANGLE
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-3,9)
x=np.zeros_like(n)
 
x[(n>=0)&(n<=4)]=n[(n>=0)&(n<=4)]
x[(n>=5)&(n<=8)]=8-n[(n>=5)&(n<=8)]
 
 
plt.stem(n,x,use_line_collection=True)
plt.xticks(n) #to display each value in x axis
plt.show()

#UNIT
import numpy as np
import matplotlib.pyplot as plt
 
n=np.arange(-2,4)
x=np.zeros_like(n)
x[n==0]=1
 
plt.stem(n,x,use_line_collection=True)
plt.show()
```
