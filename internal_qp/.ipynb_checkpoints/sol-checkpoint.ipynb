9. 
Write a python function to compute discrete Fourier transform (DFT) of an N-point signal. Verify
Parseval’s theorem



with the help of a python program for the signal **x[n] = {1, 1 + j, −1, 1 − j}.**

```python
import numpy as np


def dft(x):
    N = len(x)
    w = np.exp(-1j*2*np.pi/N)
    D = np.zeros((N,N),dtype=np.csingle)
    for i in np.arange(N):
        for j in np.arange(N):
            D[i,j] = w**(i*j)
    return np.round(D@(x.T))

x = np.array([1,1+1j,-1,1-1j])
X = dft(x)
 
lhs = 0
rhs = 0

for n in np.arange(len(x)):
    lhs += np.absolute(x[n])**2
    rhs += np.absolute(X[n])**2
print(lhs,rhs/len(x))
```

1. 
Implement convolution of two sequences of length K and M (K ≥ M) using python. (Do not use
available library function for convolution.). Use the program to compute [1 2 3] ∗ [1 1 1 1].


```python
import numpy as np

def convolve(a,b):
    k = len(a)
    m = len(b)
    if k<m:
        print("input longer sequnce first")
        exit(1)
    y = np.zeros(k+m-1)
    
    for i in np.arange(k+m-1):
        if i<m:
            y[i]= np.dot(a[:i+1],b[m-i-1::-1])
        if i<k:
            y[i]= np.dot(a[i-m+1:i+1],b[::-1])
        else :
            y[i]= np.dot(a[i-m+1:],b[::-1])
    return y

y = np.convolve(np.array([1,1,1,1]),np.array([1,2,3]))
print(y)
```
